```python
ANS.1


def get_grade(percentage):
    if percentage > 90:
        return "A"
    elif 80 < percentage <= 90:
        return "B"
    elif 60 <= percentage <= 80:
        return "C"
    else:
        return "D"

try:
    percentage = float(input("Enter the percentage: "))
    if 0 <= percentage <= 100:
        grade = get_grade(percentage)
        print(f"Grade: {grade}")
    else:
        print("Invalid percentage. Please enter a percentage between 0 and 100.")
except ValueError:
    print("Invalid input. Please enter a valid percentage.")

    
    
    
    
    ANS.2
    
    
    
    def calculate_road_tax(cost_price):
    if cost_price > 100000:
        return cost_price * 0.15
    elif 50000 < cost_price <= 100000:
        return cost_price * 0.10
    elif cost_price <= 50000:
        return cost_price * 0.05

try:
    cost_price = float(input("Enter the cost price of the bike (in Rs): "))
    if cost_price >= 0:
        road_tax = calculate_road_tax(cost_price)
        print(f"Road Tax to be paid: Rs. {road_tax:.2f}")
    else:
        print("Invalid cost price. Please enter a non-negative value.")
except ValueError:
    print("Invalid input. Please enter a valid cost price.")

    
    ANS.3
    
    
    # Dictionary to store cities and their monuments
monuments_data = {
    "Delhi": ["Red Fort", "Taj Mahal"],
    "Agra": ["Taj Mahal"],
    "Jaipur": ["Jal Mahal"]
}

try:
    city = input("Enter the name of a city: ")
    city = city.capitalize()  # Capitalize the first letter for consistent matching

    if city in monuments_data:
        monuments = monuments_data[city]
        print(f"Monuments in {city}:")
        for monument in monuments:
            print(monument)
    else:
        print(f"Monuments for {city} are not available in our database.")
except KeyboardInterrupt:
    print("\nUser interrupted the program.")

    
    
    ANS.4
    
    
    def count_divisions_by_3(number):
    count = 0
    while number > 10:
        number /= 3
        count += 1
    return count

try:
    given_number = float(input("Enter a number: "))
    divisions = count_divisions_by_3(given_number)
    print(f"The number can be divided by 3 {divisions} times before it becomes less than or equal to 10.")
except ValueError:
    print("Invalid input. Please enter a valid number.")

    
    
    
    ANS.5
    
    
    
    The while loop in Python is used to repeatedly execute a block of code as long as a certain condition is true. It allows you to perform iterative tasks and keep executing the code until the condition becomes false. The while loop is beneficial when you need to repeat a set of actions based on a condition that cannot be determined beforehand or when the number of iterations is not fixed.

The general syntax of the while loop is as follows:
    while condition:
    # Code block to be executed repeatedly

    
    Here's a detailed description of when and why to use the while loop with an example:

Use Case: Input Validation
One common scenario for using a while loop is input validation. Suppose you want to take input from the user until they enter a valid value. You can use a while loop to keep asking for input until the user provides valid data.

Example:
    
    while True:
    try:
        age = int(input("Enter your age: "))
        if age >= 0 and age <= 120:
            break  # Break the loop if the input is valid
        else:
            print("Please enter a valid age between 0 and 120.")
    except ValueError:
        print("Invalid input. Please enter a valid integer.")

        
        
         this example, the while True loop runs indefinitely until the user enters a valid age between 0 and 120. If the user enters an invalid age (not an integer or outside the specified range), the loop continues to prompt for input until a valid age is provided.

Use Case: Game Loop
Another common use case for the while loop is creating game loops or interactive programs where you want to keep running the game logic or user interactions until a specific condition (e.g., game over) is met.


game_over = False
score = 0

while not game_over:
    # Game logic and user interactions here
    answer = input("Guess the word: ")
    if answer == "python":
        print("Correct!")
        score += 10
    else:
        print("Wrong answer. Try again!")
        score -= 5

    if score < 0:
        game_over = True

print("Game Over. Your final score is:", score)


In this example, the game loop continues to run until the user's score goes below zero. The user keeps guessing words, and their score is updated based on their answers. When the score becomes negative, the while loop exits, and the game is over.

Note: It is essential to ensure that the condition within the while loop eventually becomes false; otherwise, the loop will run indefinitely, resulting in an infinite loop, which could freeze the program or lead to other issues. Always ensure that the condition has a mechanism to change or exit the loop eventually.





ANS.6




row = 1
while row <= 5:
    col = 1
    while col <= row:
        print("*", end=" ")
        col += 1
    print()
    row += 1

    
    Pattern 2: Inverted Right Triangle
    
* * * * *
* * * *
* * *
* *
*

row = 5
while row >= 1:
    col = 1
    while col <= row:
        print("*", end=" ")
        col += 1
    print()
    row -= 1

    
    
    
    
    ANS.7
    
    
    Sure! To reverse the while loop and display numbers from 10 to 1, you can modify the loop's initialization, condition, and update steps. Here's how you can do it:

python


num = 10
while num >= 1:
    print(num)
    num -= 1

    
    In this code, we start with num initialized to 10. The while loop runs as long as num is greater than or equal to 1. Inside the loop, we print the value of num and then decrement num by 1 in each iteration. This way, the loop will display numbers from 10 to 1 in reverse order.
    
    
    
    ANS.8
    
    
    ure! To display numbers from 10 to 1 in reverse order using a while loop, you can modify the loop's initialization, condition, and update steps. Here's the code
    num = 10
while num >= 1:
    print(num)
    num -= 1

    In this code, we start with the variable num initialized to 10. The while loop runs as long as num is greater than or equal to 1. Inside the loop, we print the value of num, and then decrement num by 1 in each iteration. This will display numbers from 10 to 1 in reverse order.


```
